# Matplotlib lesson notebook

```python
import numpy as np
import math
import matplotlib.pyplot as plt

%matplotlib inline
```

## Line charts

Limits - plt.xlim, plt.ylim  
Color - c argument  
Alpha - transparency  
Multiple plots  

```python
x = np.array(range(1,11))
x
```

```python
y = x ** 2
```

```python
plt.plot(y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlim(5,10)
```

```python
plt.plot(x,y, c = 'aquamarine')
plt.xlim(5,10)
plt.ylim(60,100)
```

```python
z = x ** 2 - 2
```

```python
plt.plot(x,y,c = "aquamarine")
plt.plot(x,z, c = "pink", alpha = .6)
plt.xlim(4,8)
```

## Scatter charts

Size - s argument

```python
plt.scatter(x,y,c = "aquamarine",s = 1000)
plt.plot(x,z, c = "pink", alpha = 1)
```

## Labels, Titles, and Axes

plt.title  
plt.xlabel  
plt.ylabel  

Write LaTeX symbols by wrapping text with dollar signs ($)

```python
plt.scatter(x,y,c = "aquamarine",s = 1000)
plt.scatter(x,z, c = "pink", alpha = 1)
plt.title("$subjective$ ns")
plt.xlabel('my ex')
plt.ylabel('my run')
```

## Custom Ticks

plt.xticks  
plt.yticks  

```python
plt.scatter(x,y,c = "aquamarine",s = 1000)
plt.scatter(x,z, c = "pink", alpha = 1)
plt.xticks(range(1,11))
```

## Custom Line Types

Line style - ls argument (accepts ":" and "- -")

```python
plt.scatter(x,y,c = "aquamarine",s = 1000)
plt.scatter(x,z, c = "pink", alpha = 1)
plt.title("$subjective$ ns")
```

## Rotating text and annotations

Rotation - rotation argument

```python
plt.scatter(x,y, ls = ':')
plt.scatter(x,z, ls = '--')
plt.text(2,10,"test", fontsize = 30)
plt.annotate(text = 'arrow', xy=(2,10),xytext=(6,60),arrowprops ={'width':10})
plt.title("$subjective$ ns")
```

## Adding text

plt.text - x and y coordinates, text, fontsize, color

plt.annotate - text, pointer coordinates, text coordinates, arrow properties

```python
plt.scatter(x,y, ls = ':')
plt.scatter(x,z, ls = '--')
plt.text(2,10,"test", fontsize = 30)

plt.title("$subjective$ ns")
```

## Figure size and legends

plt.figure - figsize argument to determine size of plot  
plt.legend - loc argument for location

```python
plt.figure(figsize = (10,5))

plt.scatter(x,y, ls = ':')
plt.scatter(x,z, ls = '--')

```
