# Use matplotlib to plot the following equation:

```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
x = np.array(range(1,50))
y = (x**2) - x + 2
```

```python
plt.plot(x,y)
plt.annotate(text = 'arrow', xy=(0,0),xytext=(6,400),arrowprops ={'width':2})
```

```python
n_rows = 1
n_cols = 2


# plot the first subplot
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y)
plt.title('x ~ y')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, z)
plt.title('x ~ z')

plt.show()
```
