# Create and label 4 separate charts for the following equations (choose a range for x that makes sense):



```python
x1 = range(1,30)

x1
```

```python
y1 = [math.sqrt(n)for n in x1]

y1
```

```python
plt.title("y = $sqrt(x)$")
plt.plot(x1,y1)

```

```python
z = [(n**3) for n in x1]
z
```

```python
plt.plot(x1,z)
plt.title("y=$x^3$")
```

```python
a = [(2**n)for n in x1]
a
```

```python
plt.plot(x1,a)
plt.title('$y = pow(2,x)$')
```

```python
b = [1/(n+1) for n in x1]
b
```

```python
plt.plot(x1,b)
plt.title("$1/(x+1)$")
```

# combine the figures you created in the last step into one large figure with 4 subplots.

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x1, y1, c = "pink")
axs[0, 0].set_title('Axis [0, 0], sqrt(x)')

axs[0, 1].plot(x1, z, c = "orange")
axs[0, 1].set_title('Axis [0, 1], x^3')

axs[1, 0].plot(x1, a, c = "aquamarine")
axs[1, 0].set_title('Axis [1, 0],2^x')

axs[1, 1].plot(x1, b, c = "red")
axs[1, 1].set_title('Axis [1, 1],1/(x+1)')


#set padding
plt.tight_layout()
```

# Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.title("$math$ $matical$ $equations$")
plt.plot(x1,y1, label = " $\sqrt{x}$ ")
plt.plot(x1,z, label = " $x^3$ ")
plt.plot(x1,a, label = " $2^x$")
plt.plot(x1,b, label = " $\dfrac{1}{x+1}$ ")

plt.ylim(0,40)
plt.xlabel("x line")
plt.ylabel("y rise")
plt.legend(loc='upper right')

```
