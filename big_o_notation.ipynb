```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
n1 = range(1,10)
```

```python
n = np.array(range(1,10))
n
```

```python
x = (0 * n) + 1
x
```

```python
y = np.log(n)
y
```

```python
z = n
z
```

```python
a = n * (np.log(n))
a
```

```python
b = n**2
b
```

```python
c = 2**n
c
```

```python
d = [math.factorial(n) for n in n1]
d
```

```python
e = n**n
e
```

```python
plt.plot(n,x,label= "O(1)")


plt.plot(n,y,label= "$O(log(n)$")


plt.plot(n,z,label= "$O(n)$")


plt.plot(n,a,label= "O($n log n$)")


plt.plot(n,b,label= "O($2^n$)")


plt.plot(n,c,label= "O($n^n$)")





plt.plot(n,d,label= "O(n!)")


plt.legend(loc='upper right')
plt.title("Big O Notation")

plt.xlabel("Elements")
plt.ylabel("Operations")


plt.ylim(1,10)
```
